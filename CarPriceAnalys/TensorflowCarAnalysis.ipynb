```python
dataFrame # we cannot use transmission because it is nonnumerical data so we will remove it
```

```python
dataFrame = dataFrame.drop("transmission", axis=1)
```

```python
y = dataFrame["price"].values
x = dataFrame.drop("price",axis=1)
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=10)#data is splited 2 as train and test in both x an y
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)#scaling data again
x_test= scaler.transform(x_test)
```

```python
x_train.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(12,activation="relu"))#added new layers with 12 neurons
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))

model.add(Dense(1))#output layer
model.compile(optimizer ="adam",loss="mse")#for compiling according to mse(mean square error)
```

```python
model.fit(x=x_train, y=y_train,validation_data=(x_test,y_test),batch_size=250,epochs=300 )
```

```python
loss = pd.DataFrame(model.history.history)
```

```python
loss.head()  
```

```python
loss.plot()
```

```python
guess = model.predict(x_test)
```

```python
guess
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mean_absolute_error(y_test,guess)
```

```python
plt.scatter(y_test,guess)
```
